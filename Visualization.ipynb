```python
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt

# 한글 폰트(Windows: Malgun Gothic)
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

```

```python
csv_path ="G:/criminal_analysis/outputs/domestic_3yr_reimprisonment_rate.csv"
df = pd.read_csv(csv_path, encoding="utf-8-sig")

# year/value를 숫자로 변환하고 정렬
df["year"] = pd.to_numeric(df["year"], errors="coerce")
df["value"] = pd.to_numeric(df["value"], errors="coerce")
df = df.dropna(subset=["year", "value"]).sort_values("year")

# 1) 막대그래프(연도별)
fig, ax = plt.subplots(figsize=(8, 4))
ax.bar(df["year"].astype(int), df["value"], color="tab:blue")
ax.set_title("국내 3년 이내 재복역률 (막대)")
ax.set_xlabel("년도")
ax.set_ylabel("재복역률(%)")
ax.set_xticks(df["year"].astype(int).tolist())
ax.grid(axis="y", alpha=0.3)
plt.show()

# 2) 추세선(라인)
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(df["year"], df["value"], marker="o", linewidth=2, color="tab:orange")
ax.set_title("국내 3년 이내 재복역률 추이 (라인)")
ax.set_xlabel("년도")
ax.set_ylabel("재복역률(%)")
ax.set_xticks(df["year"].astype(int).tolist())
ax.grid(True, alpha=0.3)
plt.show()
```
